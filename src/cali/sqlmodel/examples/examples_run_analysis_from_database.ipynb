# Run from Database Example

```python
from datetime import datetime

from cali.analysis import AnalysisRunner
from cali.sqlmodel import (
    AnalysisSettings,
    load_experiment_from_database,
    print_experiment_tree,
)
```

```python
database_path = "/Users/fdrgsp/Desktop/cali_test/TSC_hSynLAM77_ACTX250730_D36_DIV54_250923_jRCaMP1b_Spt.tensorstore.zarr.db"
exp = load_experiment_from_database(database_path)
print_experiment_tree(exp, max_level="fov")
```

```python
# initialize analysis runner
analysis = AnalysisRunner()
# set experiment with new analysis settings
analysis.set_experiment(exp)
```

```python
# set new analysis settings
new_analysis_settings = AnalysisSettings(
    experiment_id=exp.id,
    created_at=datetime(2025, 11, 8, 20, 25, 1, 911683),
    threads=5,
    neuropil_inner_radius=0,
    neuropil_min_pixels=0,
    neuropil_correction_factor=0.0,
    dff_window=30,
    decay_constant=0.0,
    peaks_height_value=3.0,
    peaks_prominence_multiplier=1.0,
    peaks_height_mode="multiplier",
    peaks_distance=2,
    calcium_sync_jitter_window=5,
    calcium_network_threshold=90.0,
    spikes_sync_cross_corr_lag=5,
    spike_threshold_value=1.0,
    spike_threshold_mode="multiplier",
    burst_threshold=30.0,
    burst_min_duration=3,
    burst_gaussian_sigma=2.0,
)
```

```python
# Clear previous analysis results. This is optional but recommended to avoid
# to have different positions analyzed with different settings.
# We can automatically clear only if the settings are different:
# NOTE: we exclude id and created_at from the comparison since they will always be
# different
current_settings = exp.analysis_settings.model_dump(exclude={"id", "created_at"})
new_settings = new_analysis_settings.model_dump(exclude={"id", "created_at"})
if current_settings != new_settings:
    analysis.clear_analysis_results()
print_experiment_tree(analysis.experiment(), max_level="fov")
```

```python
# set the position to analyze
analysis.set_positions_to_analyze([0])
```

```python
analysis.run()
```

```python
# get the experiment with updated analysis results
exp = analysis.experiment()
print_experiment_tree(exp, max_level="fov")
```
